Here, making a dictionary for mapping a point in a circle based ona calendar day. Basically, 01/01 will have 1/366th (it was a leap year) of 2pi, or 360 degrees. This will be used later to make a polar plot for a calendar representation.

```python
# Make a day of year to theta mapping:
eachDay = (2*(math.pi))/366
#print(eachDay)
cale = dict()
cal = calendar.Calendar()
count=0;
for month in range(1,13):
    for day in cal.itermonthdays(1984,month):
        #print(month, " ", day, "\n")
        #if(day==0):
            #print("day is zero")
        if(day!=0):
            if(day < 10 and month < 10):
                #print('adding to 0%d0%d'%(month,day))
                cale['0%d0%d'%(month,day)] = eachDay*count;
                count= count+1;
            elif(day < 10 and month > 9):
                #print('adding to %d0%d'%(month,day))
                cale['%d0%d'%(month,day)] = eachDay * count;
                count=count + 1;
            elif(month < 10):
                #print('adding to 0%d%d'%(month,day))
                cale['0%d%d'%(month,day)] = eachDay * count;
                count = count + 1;
            else:
                #print('adding to %d%d'%(month,day))
                cale['%d%d'%(month,day)] = eachDay * count;
                count=count+1;
            
            
            

#cale
```

```python
# Plotting histogram of temperature distribution
plt.hist(paris_df['   TEMP'],bins=100)

# Separating out YEARMODA column into Year, MoDa (month and date),
# Mo (month only), Da (date only)
paris_df['YEAR']=[str(ymd)[:4] for ymd in paris_df[' YEARMODA']]
paris_df['MoDa']=[str(ymd)[4:] for ymd in paris_df[' YEARMODA']]
paris_df['MO']=[str(ymd)[4:6] for ymd in paris_df[' YEARMODA']]
paris_df['DA']=[str(ymd)[6:] for ymd in paris_df[' YEARMODA']]
```

```python
# Here, creating a dictionary where each entry is a year, and it maps
# to the dataframe with the data corresponding to the year
years = np.linspace(1984,2018,35, dtype=int)
paris_data = dict()
for year in years:
    paris_data[year] = paris_df.loc[paris_df['YEAR']=='%s'%year]
```

```python
# Create a circular plot for the daily tempeartures from 1948-2018
# Inner circle is earlier years

count=0
colors=['b','g','r','c','m','y','k','w']
circle=np.linspace(0,(2*math.pi), 360)
r = 0
fig=plt.figure(figsize=[15,15])

#ax = plt.subplot(111, polar=True)
for data in paris_data:
    r = r + 15
    # Here, data represents the year
    #print(len(paris_data[data]['MoDa']))
    for index in range(len(paris_data[data]['MoDa'])):
        angle = cale['%s'%(paris_data[data]['MoDa'][index])]
        #print(paris_data[data]['MoDa'][index]," ",angle)
        plt.polar(angle, (paris_data[data]['   TEMP'][index]+r), marker='.',c=colors[count%8])#,alpha=.2)
    line=[r+50 for n in range(360)]
    plt.polar(circle, line, c=colors[count%8], lw=1)
    count = count + 1
    #if(count == 5):
     #   break
plt.title('Average Temperature of each calendar day (1948-2018)')
        
```

```python
# Let's get the variance and averages over each year for temperature and plot that
years = []
variances = []
means = []

for data in paris_data:
    year = data
    years.append(year)
    plt.hist(paris_data[data]['   TEMP'])
    variance = paris_data[data]['   TEMP'].var()
    variances.append(variance)
    mean = paris_data[data]['   TEMP'].mean()
    means.append(mean)


fig=plt.figure()
plt.scatter(years, variances)
#plt.xticks(np.linspace(1984,2018,34))

fig=plt.figure()
plt.plot(years, means)
```
